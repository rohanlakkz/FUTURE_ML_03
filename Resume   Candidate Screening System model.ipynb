```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(
    df['Cleaned_Resume'].tolist() + [job_cleaned]
)
```

```python
job_vector = tfidf_matrix[-1]

resume_vectors = tfidf_matrix[:-1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(resume_vectors, job_vector)
```

```python
df['Score'] = similarity
```

```python
skills_list = [

"python",
"machine learning",
"deep learning",
"data analysis",
"pandas",
"numpy",
"scikit-learn",
"sql",
"tensorflow",
"keras",
"statistics",
"excel",
"power bi",
"tableau"

]
```

```python
def extract_skills(text):

    found_skills = []

    for skill in skills_list:

        if skill in text:
            found_skills.append(skill)

    return found_skills


df['Skills'] = df['Cleaned_Resume'].apply(extract_skills)

df[['Category','Skills']].head()
```

```python
job_skills = extract_skills(job_cleaned)

print(job_skills)
```
